```python
def iter_iops_filenames(root_dir="./data/IOPS"):
    root = Path(root_dir)
    for train_file in sorted(root.glob("*.train.out")):
        base = train_file.name.replace(".train.out", "")
        # only keep filenames that also have a matching test file
        if (root / f"{base}.test.out").exists():
            yield base
```

```python
all_splits = {}
for fname in iter_iops_filenames("data/IOPS"):
    # train + validation (90/10 time split)
    train_ds, val_ds = load_iops(
        group="train",
        filename=fname,
        root_dir="./data",
        validation=True,
    )  # type: ignore

    # test
    test_ds = load_iops(
        group="test",
        filename=fname,
        root_dir="./data",
    )

    all_splits[fname] = {
        "train": train_ds,
        "val": val_ds,
        "test": test_ds,
    }
```
